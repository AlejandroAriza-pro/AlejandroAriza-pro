```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

```

Ejercicio 1: Captura una noticia

```python
url = "https://www.marca.com/motor/motogp/gp-jerez/2025/04/21/horarios-gp-espana-motogp-fechas-ver-tv-carrera-circuito-jerez-angel-nieto.html"
try:
  page = requests.get(url)
except:
  print("Error al abrir URL")
```

```python
page.text
```

```python
# Parseamos el html usando BeautifulSoup y lo guardamos en la variable 'soup'
soup = BeautifulSoup(page.text, "html.parser")
soup
```

```python
# Extrameos todos los parraofs de la noticia
content = soup.find("div", "ue-c-article__body")

article = []
for p in content.find_all("p"):
  article.append(p.text)

print("\n".join(article))
```

```python
# Extraemos todos los enlaces de la noticia
content = soup.find("div", "ue-c-article__body")
links = content.find_all("a")
for link in links:
  #print(link.get("href"))
  print(f"Texto: {link.text}, Href: {link['href']}")
```

Ejemplo 2: Datos Wikipedia - Provincias de España

```python
r = requests.get("https://es.wikipedia.org/wiki/Provincia_(Espa%C3%B1a)#Tabla_de_provincias")
soup = BeautifulSoup(r.text, "html.parser")
```

```python
tabla = soup.find("table", class_="wikitable")
```

```python
# todas las columnas de la tabla
provincias = tabla.find_all("tr")
```

```python
# los datos de la primera celda de cada columna
print(provincias[1].td.prettify())
```

```python
provincias[1].td.find_all("a")
```

```python
# mostrar enlaces
enlaces = []
for p in provincias:
  if p.td:
    refs = p.td.find_all("a")
    for r in refs:
      if r.has_attr("title"):
        enlaces.append(r)
enlaces
```

```python
# dataFrame
enlaces_df = pd.DataFrame({"provincia": [e.attrs["title"] for e in enlaces],
                           "enlace": [e.attrs["href"] for e in enlaces]})
enlaces_df
```

```python
# creamos la sopa para la primera provincia
r = requests.get("http://es.wikipedia.org"+enlaces_df["enlace"][0])
soup = BeautifulSoup(r.text, "html.parser")
```

```python
# extraemos un dataframe de la barra lateral derecha
tabla = soup.find("table", class_="infobox")
```

```python
# extraemos los datos - valor de la tabla
dato, valor = [], []
for t in tabla.find_all("tr"):
  if t.th:
    if t.td:
      dato.append(t.th.text)
      valor.append(' '.join([text for text in t.td.stripped_strings]))
```

```python
datos = pd.DataFrame({"dato": dato, "valor": valor})
datos
```
